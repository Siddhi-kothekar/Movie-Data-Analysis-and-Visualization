```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset (replace 'your_dataset.csv' with the actual filename)
df = pd.read_csv('moviedata.csv')

# Display the first few rows of the dataset
print(df.head())
```

```python

# Basic statistics about the dataset
print(df.describe())
```

```python
# Top directors based on average ratings
top_directors = df.groupby('Director')['Rating'].mean().sort_values(ascending=False).head(10)
print('\nTop Directors based on Average Ratings:')
print(top_directors)
```

```python
#genre
genre_counts = df['Genre'].value_counts()
plt.figure(figsize=(30, 15))
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette='viridis')
plt.title('Genre Distribution')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=90, ha='right')
plt.show()
```

```python
# Relationship between Ratings and Revenue
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Rating', y='Revenue (Millions)', data=df, color='coral')
plt.title('Relationship between Ratings and Revenue')
plt.xlabel('Ratings')
plt.ylabel('Revenue (Millions)')
plt.show()
```

```python
#top rating movies
top_rated_movies = df.nlargest(10, 'Rating')
print("\nTop-Rated Movies:")
print(top_rated_movies[['Title', 'Rating']])
plt.figure(figsize=(12, 6))
sns.barplot(x='Rating', y='Title', data=top_rated_movies, palette='muted')
plt.title('Top-Rated Movies')
plt.xlabel('Ratings')
plt.ylabel('Movie Title')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named df

movie_counts = df['Year'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
bars = plt.bar(movie_counts.index, movie_counts.values, color='skyblue')
plt.xlabel('Year')

for index, value in enumerate(movie_counts.values):
    plt.text(bars[index].get_x() + bars[index].get_width() / 2, value, str(value),
             ha='center', va='bottom')
# Add values inside the bars
for bar in bars:

    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),bar.get_height(),  ha='center', va='bottom')

plt.ylabel('Number of Movies Released')
plt.title('Number of Movies Released Each Year')
plt.show()
```
